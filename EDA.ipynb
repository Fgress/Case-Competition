# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
print(df.head())
```

## Summary statisics

```python
summary_statistics = df.drop(columns=['ID']).describe()
print(summary_statistics)
```

## Boxplot

```python
sns.set(style="whitegrid")

# Select columns
columns_to_plot = df.columns.drop(['ID', 'Customer_Churn'])

# Plot boxplots for each column
plt.figure(figsize=(15, 10))
for i, col in enumerate(columns_to_plot):
    plt.subplot(len(columns_to_plot) // 2 + 1, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

## Frequency table (count)

```python
# Select columns
columns_to_count = df.columns.drop(['ID', 'Age'])

# Loop through each column
for column in columns_to_count:
    print(f"Frequency table for {column}:")
    print(df[column].value_counts().to_frame())
    print("\n")
```

## Correlation matrix

```python
# Set the style of the visualization
sns.set(style="white")

# Compute the correlation matrix, excluding the 'ID' column
corr = df.drop(columns=['ID']).corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=20)

# Show the plot
plt.show()
```
